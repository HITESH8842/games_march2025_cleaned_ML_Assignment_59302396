```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset (works both local + Kaggle)
try:
    df = pd.read_csv('games_march2025_cleaned.csv')
except:
    df = pd.read_csv('/kaggle/input/steam-games-dataset/games_march2025_cleaned.csv')

df = df.drop_duplicates()

rq1 = df[['price', 'average_playtime_forever', 'pct_pos_total']].dropna()

rq1['price_group'] = pd.cut(
    rq1['price'],
    bins=[-1, 0, 10, 30, 60, 1000],
    labels=['Free', 'Low', 'Medium', 'High', 'Very High']
)


corr_table = rq1[['price','average_playtime_forever','pct_pos_total']].corr()
corr_table.to_csv('RQ1_Correlation_Table.csv')

# Convert to PNG
plt.figure()
plt.axis('off')
plt.title('Correlation Table')

table = plt.table(
    cellText=np.round(corr_table.values, 3),
    colLabels=corr_table.columns,
    rowLabels=corr_table.index,
    loc='center'
)

table.scale(1, 2)
plt.savefig('RQ1_Correlation_Table.png', bbox_inches='tight')
plt.show()

print("Correlation Table:")
print(corr_table)


group_table = rq1.groupby('price_group')[['average_playtime_forever','pct_pos_total']].mean()
group_table.to_csv('RQ1_Group_Table.csv')

# Convert to PNG
plt.figure()
plt.axis('off')
plt.title('Price Group Summary')

table = plt.table(
    cellText=np.round(group_table.values, 2),
    colLabels=group_table.columns,
    rowLabels=group_table.index,
    loc='center'
)

table.scale(1, 2)
plt.savefig('RQ1_Group_Table.png', bbox_inches='tight')
plt.show()

print("\nPrice Group Summary:")
print(group_table)


plt.figure()
plt.scatter(rq1['price'], rq1['average_playtime_forever'])
plt.xlabel('Price')
plt.ylabel('Average Playtime')
plt.title('Price vs Engagement')
plt.savefig('RQ1_Price_vs_Playtime.PNG')
plt.show()


plt.figure()
plt.scatter(rq1['price'], rq1['pct_pos_total'])
plt.xlabel('Price')
plt.ylabel('Positive Reviews (%)')
plt.title('Price vs Satisfaction')
plt.savefig('RQ1_Price_vs_Satisfaction.PNG')
plt.show()


plt.figure()
group_table['average_playtime_forever'].plot(kind='bar')
plt.title('Average Playtime by Price Group')
plt.ylabel('Playtime')
plt.savefig('RQ1_Playtime_by_PriceGroup.PNG')
plt.show()


plt.figure()
group_table['pct_pos_total'].plot(kind='bar')
plt.title('Satisfaction by Price Group')
plt.ylabel('Positive %')
plt.savefig('RQ1_Satisfaction_by_PriceGroup.PNG')
plt.show()

```
