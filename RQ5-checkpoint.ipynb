```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

os.makedirs('Tables', exist_ok=True)
os.makedirs('Models', exist_ok=True)

# ================= LOAD DATA =================
df = pd.read_csv(r'games_march2025_cleaned.csv')
df = df.drop_duplicates()

# ================= CLEAN =================
df['peak_ccu'] = pd.to_numeric(df['peak_ccu'], errors='coerce')
df['average_playtime_2weeks'] = pd.to_numeric(df['average_playtime_2weeks'], errors='coerce')
df['num_reviews_total'] = pd.to_numeric(df['num_reviews_total'], errors='coerce')


rq5 = df[['categories','peak_ccu','average_playtime_2weeks','num_reviews_total']].dropna()

rq5['multiplayer'] = rq5['categories'].str.contains('Multi', case=False, na=False)

# ================= GROUPING =================
grouped_mean = rq5.groupby('multiplayer')[['peak_ccu','average_playtime_2weeks','num_reviews_total']].mean()
grouped_median = rq5.groupby('multiplayer')[['peak_ccu','average_playtime_2weeks','num_reviews_total']].median()
grouped_count = rq5.groupby('multiplayer').size()

# FIX labels safely
labels = ['Single', 'Multiplayer']

# ================= TABLE 1: MEAN =================
plt.figure()
plt.axis('off')

plt.table(
    cellText=np.round(grouped_mean.values, 2),
    colLabels=grouped_mean.columns,
    rowLabels=labels,
    loc='center'
)

plt.title('RQ5 Table 1: Mean Comparison')
plt.savefig('Tables/RQ5_Table1_Mean.png', dpi=300, bbox_inches='tight')
plt.show()

# ================= TABLE 2: MEDIAN =================
plt.figure()
plt.axis('off')

plt.table(
    cellText=np.round(grouped_median.values, 2),
    colLabels=grouped_median.columns,
    rowLabels=labels,
    loc='center'
)

plt.title('RQ5 Table 2: Median Comparison')
plt.savefig('Tables/RQ5_Table2_Median.png', dpi=300, bbox_inches='tight')
plt.show()

# ================= TABLE 3: COUNT =================
plt.figure()
plt.axis('off')

plt.table(
    cellText=grouped_count.values.reshape(-1,1),
    colLabels=['Count'],
    rowLabels=labels,
    loc='center'
)

plt.title('RQ5 Table 3: Sample Size')
plt.savefig('Tables/RQ5_Table3_Count.png', dpi=300, bbox_inches='tight')
plt.show()

# ================= FIGURE 1 =================
plt.figure()
grouped_mean['peak_ccu'].plot(kind='bar')
plt.title('Peak CCU Comparison')
plt.savefig('Models/RQ5_Fig1_Peak.png', dpi=300, bbox_inches='tight')
plt.show()

# ================= FIGURE 2 =================
plt.figure()
grouped_mean['average_playtime_2weeks'].plot(kind='bar')
plt.title('Playtime Comparison')
plt.savefig('Models/RQ5_Fig2_Playtime.png', dpi=300, bbox_inches='tight')
plt.show()

# ================= FIGURE 3 =================
plt.figure()

plt.boxplot(
    [
        rq5[rq5['multiplayer'] == False]['peak_ccu'],
        rq5[rq5['multiplayer'] == True]['peak_ccu']
    ],
    labels=['Single', 'Multiplayer']
)

plt.title('Distribution of Peak CCU')
plt.savefig('Models/RQ5_Fig3_Boxplot.png', dpi=300, bbox_inches='tight')
plt.show()
```
